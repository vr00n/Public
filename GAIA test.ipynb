# Extract Gaia Articles

Using Google Chrome's Developer Mode, I was able to locate the service that served Gaia articles. In this cell, I iterate through each 10 pages of content, each set to contain 100 articles and save it as a JSON file.


```python
import os
import sys

for i in range (1,11):
    full_cmd="curl 'https://d6api.gaia.com/articles?p="+str(i)+"&pp=100' -H 'pragma: no-cache' -H 'origin: https://www.gaia.com' -H 'accept-encoding: gzip, deflate, br' -H 'accept-language: en-US,en;q=0.8' -H 'user-agent: Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36' -H 'content-type: application/json' -H 'accept: application/json' -H 'cache-control: no-cache' -H 'authority: d6api.gaia.com' -H 'referer: https://www.gaia.com/' --compressed >>GAIA-articles-"+str(i)+".json"
    os.system(full_cmd)
```

# For each article, we retrieve the Article Body and compute a Sentiment Score

I was also able to locate the service that contains the full-text of each article on Gaia.com
For each set of Articles contained in the previously extracted JSON files, I retrieve the Article Body and then pass it to a Sentiment Analysis method to compute a simple Sentiment Analysis Score.

```python
import json
from pprint import pprint
import requests
import urllib
from textblob import TextBlob
import glob
x_=[]
y_=[]
text_=[]
article_text=[]
sentiment_polarity=[]

for filename in glob.iglob('*.json'):
    with open(filename) as data_file:    
        data = json.load(data_file)

    for i in data['articles']:
        up_vote=i['fivestar']['up_count']['value']
        total_votes=(i['fivestar']['up_count']['value']+i['fivestar']['down_count']['value'])
        full_path="https://brooklyn.gaia.com/node/"+i['nid']+"?f=md"
        response = urllib.urlopen(full_path)
        article_text = json.loads(response.read())
        article_len = len(article_text['fields']['body'][0]['value'])
        article_body= TextBlob(article_text['fields']['body'][0]['value'])
        article_sent_pol=article_body.sentiment.polarity

        if total_votes >0:
            perc_up = round(float(up_vote)/float(total_votes)*100,2)
        else:
            perc_up=0
        x_.append(float(perc_up))
        y_.append(float(article_len))
        text_.append(i['title'])
        sentiment_polarity.append(article_sent_pol)
        
        print i['nid'],article_sent_pol,article_len        
```



# Plotting Sentiment Score against Article length

For each Gaia Article, I plot Length of Article Body vs Sentiment Score using the Plotly service.

```python
import plotly.plotly as py
import plotly.graph_objs as go

trace1 = go.Scatter(x=y_, y=sentiment_polarity, marker={'color': 'red', 'size': "6"}, 
                    mode="markers",  text=text_, name='1st Trace')
                                               
data=go.Data([trace1])
layout=go.Layout(title="First Gaia Plot", xaxis={'title':'Article Length'}, yaxis={'title':'Sentiment Score'})
figure=go.Figure(data=data,layout=layout)
py.iplot(figure, filename='pyguide_1')
```
